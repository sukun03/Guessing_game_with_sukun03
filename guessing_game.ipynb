```python
import random
import time

print("HELLO SIR IF YOU WANT TO PLAY GAME TYPE 'Y' IF NOT TYPE 'N'")
user = input("y or n: ")

if user == "y":

    def game():
        num = random.randrange(1, 100)


        total_time = 20
        start_time = time.time()

        while True:
            elapsed = int(time.time() - start_time)
            remaining = total_time - elapsed

            if remaining <= 0:
                print("Time over! You lost the game.")
                print("Correct number was:", num)
                break

            print("⏳ Time left:", remaining, "seconds")

            try:
                guess = int(input("Enter a number: "))

                if guess == num:
                    print(" Congratulations! You got the number")
                    break
                elif guess > num:
                    print("Guess lower number")
                else:
                    print("Guess greater number")
            except ValueError:
                print("Invalid input. Please enter a valid integer.")


    game()

else:
    print("NICE TO MEET U BYE")
```
